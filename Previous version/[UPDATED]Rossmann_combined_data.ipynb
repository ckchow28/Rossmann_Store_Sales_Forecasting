```python
import pandas as pd


sales_df = pd.read_csv("data/Rossmann_store.csv", low_memory = False)
store_df = pd.read_csv("data/Rossmann_train.csv", low_memory = False)

combined_df = pd.merge(sales_df, store_df, on='Store')
combined = combined_df.dropna()
```

```python
combined_sorted = combined.sort_values(by='Sales', ascending = False)
combined_sorted 
```

```python
top_20_percent_index = int(len(combined_sorted) * 0.2)
top_20_percent_sales = combined_sorted.head(top_20_percent_index)
top_20_percent_sales
```
